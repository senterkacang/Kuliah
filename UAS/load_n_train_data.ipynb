Load dataset dari sklearn dataset

```python
import random
import numpy as np
from sklearn.datasets import load_iris
from network import Network


# 1. Load dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# 2. Bentuk vektor input (4x1)
X = [x.reshape(4,1) for x in X]

# 3. One-hot output
def one_hot(k):
    v = np.zeros((3,1))
    v[k] = 1.0
    return v

y_vec = [one_hot(k) for k in y]

# 4. Gabungkan dataset
data = list(zip(X, y_vec))

# 5. Shuffle & split (120 train, 30 test)
random.shuffle(data)
train_data = data[:120]
test_data = [(x, np.argmax(yv)) for x, yv in data[120:]]
```

Tampilkan isi dalam dataset tersebut

```python
import pandas as pd

iris = load_iris()
df = pd.DataFrame(
    iris.data,
    columns=["length1", "width1", "length2", "width2"]
)
df["species"] = iris.target   # 0,1,2

print(df)
```

Lakukan training dataset

```python
net = Network([4, 20, 20, 3])
net.train(train_data, epochs=600, mini_batch_size=10, eta=0.1, test_data=test_data)

```

Save hasil training

```python
net.save("iris_nn_model.pkl")
```
